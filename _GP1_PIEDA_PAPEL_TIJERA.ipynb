```python
import random

reglas_victoria = {
    'tijera': ['papel', 'lagarto'],
    'papel': ['piedra', 'spock'],
    'piedra': ['lagarto', 'tijera'],
    'lagarto': ['spock', 'papel'],
    'spock': ['tijera', 'piedra']
}

def determinar_ganador(jugador1, jugador2, eleccion1, eleccion2):
    """
    Función que determina el resultado de una ronda entre dos jugadores.
    """
    if eleccion1 == eleccion2:
        return 'empate'
    elif eleccion2 in reglas_victoria[eleccion1]:
        return jugador1
    else:
        return jugador2

def jugar_ronda(jugadores):
    """
    Función para jugar una ronda del juego.
    """
    print("\n--- ¡Comienza la ronda! ---")

    elecciones = {}

    for jugador in jugadores:
        while True:
            opcion = input(f"{jugador}, elige tu opción (piedra, papel, tijera, lagarto, spock): ").lower()
            if opcion in reglas_victoria:
                elecciones[jugador] = opcion
                break
            else:
                print("Opción no válida. Por favor, elige entre piedra, papel, tijera, lagarto o spock.")

    print("\n--- Resultados de la ronda ---")
    for jugador, eleccion in elecciones.items():
        print(f"{jugador} eligió: {eleccion}")

    puntos_ronda = {}
    for jugador in jugadores:
        puntos_ronda[jugador] = 0

    if len(jugadores) > 1:
        resultados = {}
        for i in range(len(jugadores)):
            for j in range(i + 1, len(jugadores)):
                jugador1 = jugadores[i]
                jugador2 = jugadores[j]
                eleccion1 = elecciones[jugador1]
                eleccion2 = elecciones[jugador2]

                ganador = determinar_ganador(jugador1, jugador2, eleccion1, eleccion2)

                if ganador != 'empate':
                    if ganador not in resultados:
                        resultados[ganador] = 0
                    resultados[ganador] += 1

        if resultados:
            ganador_ronda = max(resultados, key=resultados.get)
            print(f"\n¡El ganador de esta ronda es: {ganador_ronda}!")
            puntos_ronda[ganador_ronda] = 1
        else:
            print("\n¡La ronda terminó en empate para todos los jugadores!")

    return puntos_ronda


def jugar_juego():
    """
    Función principal para gestionar el flujo del juego completo.
    """
    print("--- ¡Bienvenido al juego de Piedra, Papel, Tijera, Lagarto, Spock! ---")
    print("Recuerda las reglas:")
    for opcion, gana_a in reglas_victoria.items():
        print(f"- {opcion.capitalize()} vence a {', '.join(gana_a)}.")
    print("----------------------------------------------------------------------")

    while True:
        try:
            num_jugadores = int(input("Ingresa el número de jugadores (de 2 a 4): "))
            if 2 <= num_jugadores <= 4:
                break
            else:
                print("Número de jugadores no válido. Por favor, ingresa un número entre 2 y 4.")
        except ValueError:
            print("Entrada no válida. Por favor, ingresa un número.")

    jugadores = []
    for i in range(num_jugadores):
        jugadores.append(input(f"Ingresa el nombre del jugador {i+1}: "))

    puntajes = {jugador: 0 for jugador in jugadores}
    rondas_jugadas = 0
    max_rondas = 3

    while rondas_jugadas < max_rondas:
        rondas_jugadas += 1
        print(f"\n==================== Rondas: {rondas_jugadas}/{max_rondas} ====================")

        puntos_ronda = jugar_ronda(jugadores)

        for jugador, puntos in puntos_ronda.items():
            puntajes[jugador] += puntos

        print("\n--- Marcador actual ---")
        for jugador, puntos in puntajes.items():
            print(f"{jugador}: {puntos} puntos")

    print("\n==================== Fin del Juego ====================")

    ganador_final = max(puntajes, key=puntajes.get)
    puntaje_ganador = puntajes[ganador_final]

    #Verificar si hay un empate en el puntaje más alto
    ganadores_finales = [jugador for jugador, puntos in puntajes.items() if puntos == puntaje_ganador]

    if len(ganadores_finales) > 1:
        print("¡El juego ha terminado en un empate!")
        print("Los jugadores empatados son:", ", ".join(ganadores_finales))
    else:
        print(f"¡El ganador final es: {ganador_final} con {puntaje_ganador} puntos!")

jugar_juego()
```
